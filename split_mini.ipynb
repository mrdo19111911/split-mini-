```python
import math
import numpy as np
import numba 
from numba import njit
```

```python
n =50
k = 5
location= np.random.randint(1, 1000000, (n,2))
# DM = np.empty((n,n, k), np.int_)
DM_veh = np.empty((n,n, k), np.int_)

for i in range(n):
    for j in range(n):
        DM_veh[i,j] =  round(math.sqrt(  ( location[i][0]-location[j][0] )**2+( location[i][1]-location[j][1] )**2)  *(k+1) ) 
        # DM_veh[i,j] =  round(1)
demand =  np.random.randint(1, 2, (n))
# DM_veh = np.empty((n,n, k), np.int_)
```

```python
@njit
def create_virtual_graph(DM,demand,capacity):
    c = np.empty((n,n), np.int_)
    for i in range(n):
        for j in range(n):
            c[i,j]=10000000000
    for i in range(n):
        cost = 0
        demand_acc = 0
        c[i,i]=0
        for j in range(i+1,n-1):
            if j==i+1:
                cost= DM[0][i+1]
            else:
                cost += DM[j-1][j]
            demand_acc+= demand[j]
            if demand_acc>capacity:
                # c[i,j]=0
                break
            else:
                c[i,j]= cost +DM[j][0]
    return c
DM =  DM_veh[1]
all_c=[]
for _ in range(k):
    all_c.append(create_virtual_graph(DM, demand,6))
# c = create_virtual_graph(DM, demand, 2000000)


# for x in c:
#     print(x)
```

```python
@njit
def find_shortest_path(c):
    price= np.zeros((n), np.int_)
    pred = np.zeros((n), np.int_)
    for i in range(1,n):
       price[i]= 100**6 
    for i in range(len(c)-1):
        j = i + 1
        # print(i,j)

        while j < len(c) and c[i][j]<100000000:
            # temp= np.zeros((n), np.int_)
            # print(price[i] ,c[i][j], price [j])

            if price[i] + c[i][j] <= price [j]:
                price[j] = price[i] + c[i][j]
                pred[j] = i
            j+=1

        temp = [ x if x < 1000000 else -1 for x in price ]
        


    return price, pred

price, pred = find_shortest_path(all_c[3])
price, pred 
```
